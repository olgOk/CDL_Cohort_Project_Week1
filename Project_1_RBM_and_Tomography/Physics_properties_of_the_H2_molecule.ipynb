# Bayesian Polynomial Fitting

We are interested in approximating the poential function $V(r)$ as a polynomial of the form

$$
V(r) = E_0 +\frac{1}{2}k\left( r - r_0 \right)^2,
$$

where the parameters in the above fit are: 
1. $E_0$ the minimum energy of the well that represents the most stable energy configuration of the two protons in the $H_2$ molecule,
2. $k$ a parameter that represents the 'stiffness' of the $H_2$ molecule,
3. $r_0$ which is the average distance between the two protons in the $H_2$ molecule. 

In order to fit our data generated by the RBM that carries some uncertainty to this function we will carry out a Bayesian fit of the model. The Bayesian fit will allow us to asses the uncertainty in the parameters of the models and help us understand the error in our predictions.

```python
# To carry out the Bayesian estimate, we will use the package EMCEE which requires us to define the log likelyhoods
# of the parameters that we are fitting

def log_likelihood(theta, x, y, yerr):
    '''
    The log-likelyhood of the model
    
    model = E0+1/2*k*(x-r0)**2
    
    '''
    E0, k, r0, log_f = theta
    
    model = E0+0.5*k*(x-r0)**2
    
    sigma2 = yerr ** 2 + model ** 2 * np.exp(2 * log_f)
    
    return -0.5 * np.sum((y - model) ** 2 / sigma2 + np.log(sigma2))

def log_prior(theta):
    '''
        The log prior of the parameters of the Bayesian model
    '''
    
    E0, k,r0, log_f = theta
    
    if -20.0 < E0 < 20.0 and -10.0 < k < 10.0 and -2.0 < r0 < 2.0 and -10.0 < log_f < 1.0:
        return 0.0
    
    return -np.inf

def log_probability(theta, x, y, yerr):
    '''
        The log-probability of the Bayesian model 
    '''
    
    lp = log_prior(theta)
    if not np.isfinite(lp):
        return -np.inf
    return lp + log_likelihood(theta, x, y, yerr)


def l2(theta,x,y):
    '''
    The L2 loss function 
    
    '''
    
    E0, k, r0 = theta

    model = E0+0.5*k*(x-r0)**2
    
    loss = np.sum((model-y)**2)
    
    return loss

def confidence_interval(x):
    '''
    
        Compute the 68% and 96% confidence inteval of a data set
        
        Args:
            x: numpy array
        Returns:
            Tuple: array1,array2
            
            where array1 and array2 are each numpy arrays of length=3
            array1 = [lower_68%_bound, median , upper_68%_bound]
            array2 = [lower_96%_bound, median , upper_96%_bound]
    '''
    x_68 = np.percentile(x, [16,50,84])
    x_96 = np.percentile(x, [2,50,98])
    
    return x_68,x_96

def convert_confidence_region_to_uncertainty(x_low,x_med,x_high,Nround=3):
    '''
        The function converts a confidence interval range into a string
        that represents the uncertainty interval.

        Reuturns:
                string:  'x0^ {+upper}_{-lower}''
    '''

    upper = str(round(abs(x_high-x_med),Nround))
    lower = str(round(abs(x_low-x_med),Nround))

    s = str(round(x_med,Nround))+" ^{ +" + upper + " }"+" _{ -" + lower + " }"

    return s
```

```python
# We use the max-likelyhood method to determine initial parameters for the MCMC sampler. Then the MCMC sampler 
# explores the log likelyhood landscape in order to determine the posterior distribution of the parameters.

nll = lambda *args: l2(*args)
initial = np.array([-1.7,8.01,6.3]) # Some not too-unreasonable initial guess
soln = minimize(nll, initial, args=(x, y))
E0_ml,k_ml,r0_ml = soln.x

pos = np.array([E0_ml,k_ml,r0_ml,-4.0]) + 1e-4 * np.random.randn(32, 4)
nwalkers, ndim = pos.shape

sampler = emcee.EnsembleSampler(nwalkers, ndim, log_probability, args=(x, y, yerr))
sampler.run_mcmc(pos, 5000, progress=True);
```

```python
# After the sampler has run, we can plot the joint-probability distribution of the parameters
# and plot the 68% confindence intervals 

flat_samples = sampler.get_chain(discard=100, thin=15, flat=True)

fig = corner.corner(
    flat_samples, labels=['E0','k','r0','f'],
    quantiles=[0.16, 0.5, 0.84],
    show_titles=True, title_kwargs={"fontsize": 12}
);
```

```python
# For completeness let us plot some samples from the sampler 

plt.figure(figsize=(10,5)) 

inds = np.random.randint(len(flat_samples), size=100)
for ind in inds:
    E0,k,r0,log_f = flat_samples[ind]
    plt.plot(x, E0+0.5*k*(x-r0)**2, color='red',alpha=0.1)
       
plt.plot(x, E0+0.5*k*(x-r0)**2, color='red',alpha=0.3,label='MCMC-samples')
#plt.fill_between(x,y-yerr,y+yerr,color='blue',alpha=0.3)
plt.errorbar(x,y,yerr=yerr,color='k', label='data')
plt.legend(fontsize=14)
plt.xlabel("radius [Angstrom]",size=15)
plt.ylabel("Energy [Hartree]",size=15);
```

# Determining the vibrational and rotational modes of the $H_2$ molecule

In the code above, we determined the posterior distributions of the several of the properties of the $H_2$ molecule. Now, we can use these properties in order to determine the vibrational and rotational modes of the molecule. But, first let us condider some physics.


Around the minimum of our potential energy surface, the dynamics of our molecule can be approximated as a harmonic oscillator
$$
H = \frac{p^2}{2\mu}+V(r) \\
H \approx \frac{p^2}{2\mu}+\frac{1}{2}k(r-r_0)^2
$$
where $\mu$ is the reduced mass of the proton-proton system, $k$ is the spring constant and $r_0$ is the most likely distance between the two protons. From the above Hamiltonian, the dynamics of the two protons will obey the equation 

$$
\frac{d^2 r(t)}{d t^2} = - \frac{k}{\mu} \left(r-r_0 \right). 
$$

This is simple harmonic motion with frequency given by

$$
\omega = \sqrt{\frac{k}{\mu}}.
$$


From the quantum mechanics of a harmonic oscillator, the vibrational frequencies of our system will be given by

$$
E_{\rm vibrational} = \hbar \omega \left( n +\frac{1}{2} \right),
$$
where $n=0,1 \cdots$ are integers. 

In the cell below, we will work out what the fundamental vibrational frequency of the $H_2$ molecule is, and compare the results with experiment.

```python
hartree_2_MeV = 2.7211386246*10**-5  # Convert Hartree unit to MeV
angstrom_2_fm = 10**5 # Convert Angstrom to fm
mass_proton = 938.27231 # MeV/c^2
reduced_mass = 0.5*mass_proton # Mev/c^2
hbarc = 197.327 # fm*MeV
fm_2_cm = 10**-13
MeV_2_eV = 10**6
eV_2_cm_m1 = 8066 # cm^-1

freq_exp = 4401 # cm^-1 Experimental value, see: https://www.cup.uni-muenchen.de/ch/compchem/vib/vib1.html



k_samples = flat_samples[:,1]*(hartree_2_MeV/angstrom_2_fm**2) # Convert to MeV/fm**2
w_samples = np.sqrt(k_samples/reduced_mass) # c/fm
hbar_w_samples = w_samples*hbarc*MeV_2_eV # eV

# The energy of fundamental mode = E=0.5*hbar*omega
freq_samples = 0.5*hbar_w_samples*eV_2_cm_m1 # 1/cm**-1


freq_68,freq_96 = confidence_interval(freq_samples)

freq_string = convert_confidence_region_to_uncertainty(freq_68[0],freq_68[1],freq_68[2],Nround=0)


plt.figure(figsize=(10,5))
height, bins, patches = plt.hist(freq_samples,bins=30)

plt.title(r'$f_0$='+freq_string+ ' [1/cm]',size=15)
plt.vlines(freq_68[0],ymin=0, ymax = np.max(height),linestyles=':')
plt.vlines(freq_68[1],ymin=0, ymax = np.max(height),linestyles='-',label= 'k-median' )
plt.vlines(freq_68[2],ymin=0, ymax = np.max(height),linestyles=':')
plt.vlines(freq_exp,ymin=0, ymax = np.max(height),linestyles='-',color='red',label='Experiment'+'\n'+'f0='+str(freq_exp)+'cm^-1')
plt.xlabel('$f_0$ [1/cm]',size=25)
#plt.xlim(0.6,1.4)
plt.legend()
plt.show()
```

We see that our median value agrees really well with the experimental values! This is very cool considering we obtained this information by training an RBM.

# Heat capacity of the $H_2$ molecular gas

Using the vibrational energy that we found earlier, we can now make some predictions about the heat capacity of the Hydrogen molecular gas. Including the uncertainty bands that arise from the indeterminate value that we found earlier.

```python
kb = 8.617333262*10**(-5)# eV K-1

def cv_vibrations(T,hbar_omega):
    '''
        The Einstein vibrational model of the heat capacity (https://en.wikipedia.org/wiki/Einstein_solid)
        
        
    
    '''
    
    x = (hbar_omega/(kb*T)) # Dimensionless
    
    cv = 3.0*kb*x**2*np.exp(-x)/(1.0-np.exp(-x))**2
    
    return cv
```

```python

hbar_w0_68,hbar_w0_96 = confidence_interval(hbar_w_samples)

T=np.linspace(1e3,1e5,3000)

plt.figure(figsize=(12,5))
plt.title('The contribution of the vibrational modes to the $H_2$ heat capacity',size=15)
plt.plot(T,cv_vibrations(T,hbar_omega=hbar_w0_68[1]),color='blue')
plt.fill_between(T,cv_vibrations(T,hbar_omega=hbar_w0_68[0]),cv_vibrations(T,hbar_omega=hbar_w0_68[2]),alpha=0.4,color='blue')
plt.fill_between(T,cv_vibrations(T,hbar_omega=hbar_w0_96[0]),cv_vibrations(T,hbar_omega=hbar_w0_96[2]),alpha=0.2,color='blue')
plt.plot(T,np.ones(len(T))*3.0*kb,color='r',label=r'$3k_b$')
plt.xlabel('Temperature [K]',size=15)
plt.ylabel('$c_v(T)$ [eV K^{-1}]',size=20)
plt.xscale('log')
plt.legend()
plt.show()
```

We observe that the vibrational modes do not become important to the heat capacity until in the intermediate values between T=$10^3$-$10^4$ K. Furthermore, the uncertainty only effects this intermediate region. The uncertainty analysis is therefore critical for applications where the temperature of the $H_2$ gas is in this region. 

# The Rotational Occupation numbers of the $H_2$ molecule

Another interesting property of the $H_2$ molecule that we can gain some insight from is the rotational occupation numbers. The quantum mechanical energy of a rigid rotor is given by the expression

$$
E_{\rm rotor} = \frac{J(J+1)}{2} \frac{\hbar^2}{I},
$$
where $I$ is the moment of inertia given by
$$
I = \mu r^2_0.
$$

```python
r0_samples = flat_samples[:,2]

I_samples = reduced_mass*(r0_samples)**2

I_68,I_96 = confidence_interval(I_samples)

I_string = convert_confidence_region_to_uncertainty(I_68[0],I_68[1],I_68[2],Nround=2)


plt.figure(figsize=(10,5))
height, bins, patches = plt.hist(I_samples,bins=30)

plt.title(r'$I$='+I_string+ ' [MeV*Angstrom**2]',size=15)
plt.vlines(I_68[0],ymin=0, ymax = np.max(height),linestyles=':')
plt.vlines(I_68[1],ymin=0, ymax = np.max(height),linestyles='-',label= 'k-median' )
plt.vlines(I_68[2],ymin=0, ymax = np.max(height),linestyles=':')
#plt.vlines(freq_exp,ymin=0, ymax = np.max(height),linestyles='-',color='red',label='Experiment'+'\n'+'f0='+str(freq_exp)+'cm^-1')
plt.xlabel('Moment of inertia $I$ [MeV*Angstrom**2]',size=20)
#plt.xlim(0.6,1.4)
plt.legend()
plt.show()
```

```python
def rotational_energy(J,I):
    '''
    
    '''
    
    # Convert I from MeV*Angstrom**2 to MeV*fm**2
    
    I = I*(angstrom_2_fm)**2
    
    EJ = 0.5*(J*(J+1.0))*(hbarc**2)/I # MeV
    
    return EJ

def occupation_number(J,I,T):
    '''
    
    '''
    
    kb_MeV_inv_K = 8.617*10**(-11)# MeV*K^(-1) (reciprocal kelvins)
    
    n = (2.0*J+1.0)*np.exp(-rotational_energy(J,I)/(kb_MeV_inv_K*T))
    
    j_vec = np.array(range(0,1000))
    
    # The partition function
    Z = np.sum((2.0*j_vec+1.0)*np.exp(-rotational_energy(j_vec,I)/(kb*T)))
    
    p = n/Z
    
    return p
```
